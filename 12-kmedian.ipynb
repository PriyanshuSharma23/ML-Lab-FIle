# K-Median Clustering Algorithm

The K-Median Clustering Algorithm is a clustering algorithm that is used to cluster data into K clusters. It is similar to the K-Means algorithm, except for the fact that the K-Means algorithm uses the mean value of the points in a cluster to determine the centroid of that cluster, whereas the K-Median algorithm uses the median value of the points in a cluster to determine the centroid of that cluster.

## Algorithm

The K-Median Clustering Algorithm is as follows:

1. Initialize K cluster centroids randomly.
2. Assign each point to the cluster with the nearest centroid.
3. Recompute the centroid of each cluster by taking the median of all the points in the cluster.
4. Repeat steps 2 and 3 until the centroids stop changing.



```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(0)
X = np.random.randn(100, 2)

# Visualize the data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.axis('equal')
plt.show()
```

```python
# Define the number of clusters (k) and maximum iterations
k = 3
max_iterations = 100
```

```python
# Initialize cluster centers randomly
initial_centers = X[np.random.choice(X.shape[0], k, replace=False)]
```

```python
# K-median clustering algorithm
for _ in range(max_iterations):
    # Assign each point to the nearest cluster center
    labels = np.argmin(np.linalg.norm(X[:, np.newaxis] - initial_centers, axis=2), axis=1)
    
    # Update cluster centers to be the median of the assigned points
    new_centers = np.array([np.median(X[labels == i], axis=0) for i in range(k)])
    
    # Check for convergence
    if np.all(initial_centers == new_centers):
        break
    
    initial_centers = new_centers
```

```python
# Visualize the clusters and cluster centers
for i in range(k):
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1,], label=f'Cluster {i + 1}')

plt.scatter(initial_centers[:, 0], initial_centers[:, 1,], color='black', marker='x', s=150, label='Cluster Centers')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-median Clustering')
plt.legend()
plt.show()
```
