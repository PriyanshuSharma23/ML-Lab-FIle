# Principal Component Analysis - PCA

The principal components of a collection of points in a real p-space that are a sequence of p direction vectors, where the i-th vector is the direction of a line that best fits the data while being orthogonal to the first i âˆ’ 1 vectors.

The principal components transformation can be defined as a transformation that maps the data to a new coordinate system such that the greatest variance by some scalar projection of the data comes to lie on the first coordinate (called the first principal component), the second greatest variance on the second coordinate, and so on.

The principal components are orthogonal because they are the eigenvectors of the covariance matrix, which is symmetric. PCA is sensitive to the relative scaling of the original variables.


```python
# importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Create a 2D dataset
np.random.seed(0)
data_2d = np.random.randn(100, 2)
```

```python
# Create a 3D dataset
np.random.seed(0)
data_3d = np.random.randn(100, 3)
```

```python
# Initialize PCA for 2D to 1D
pca_2d_to_1d = PCA(n_components=1)
# Fit and transform the 2D data
data_1d = pca_2d_to_1d.fit_transform(data_2d)
```

```python
# Initialize PCA for 3D to 2D
pca_3d_to_2d = PCA(n_components=2)
# Fit and transform the 3D data
data_2d_from_3d = pca_3d_to_2d.fit_transform(data_3d)
```

```python
# Visualize the data before and after PCA
plt.figure(figsize=(12, 5))

# Plot 2D data
plt.subplot(1, 3, 1)
plt.scatter(data_2d[:, 0], data_2d[:, 1])
plt.title("Original 2D Data")

# Plot 1D data after PCA
plt.subplot(1, 3, 2)
plt.scatter(data_1d, np.zeros_like(data_1d))
plt.title("Reduced to 1D")
```

```python
# Plot 3D data
plt.subplot(1, 3, 3)
plt.scatter(data_3d[:, 0], data_3d[:, 1], c=data_3d[:, 2])
plt.title("Original 3D Data")

plt.show()

# Visualize the 2D data after PCA
plt.figure(figsize=(6, 5))
plt.scatter(data_2d_from_3d[:, 0], data_2d_from_3d[:, 1])
plt.title("Reduced to 2D from 3D using PCA")

plt.show()
```

```python
# Plot the principal components for 2D to 1D
plt.figure()
plt.quiver(0, 0, pca_2d_to_1d.components_[0, 0], pca_2d_to_1d.components_[0, 1], angles='xy', scale_units='xy', scale=1, color='r')
plt.title("Principal Component for 2D to 1D PCA")
```

```python
# Plot the principal components for 3D to 2D
plt.figure()
plt.quiver(0, 0, pca_3d_to_2d.components_[0, 0], pca_3d_to_2d.components_[0, 1], angles='xy', scale_units='xy', scale=1, color='r', label="PC1")
plt.quiver(0, 0, pca_3d_to_2d.components_[1, 0], pca_3d_to_2d.components_[1, 1], angles='xy', scale_units='xy', scale=1, color='b', label="PC2")
plt.legend()
plt.title("Principal Components for 3D to 2D PCA")

plt.show()
```
