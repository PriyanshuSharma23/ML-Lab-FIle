# Regression

Regression is a statistical approach to find the relationship between variables. In machine learning, this is used to predict the outcome of an event based on the relationship between variables obtained from the data-set. More specifically, regression analysis helps us to understand how the value of the dependent variable is changing corresponding to an independent variable when other independent variables are held fixed.

## Linear Regression

Linear regression is a basic and commonly used type of predictive analysis.  The overall idea of regression is to examine two things: (1) does a set of predictor variables do a good job in predicting an outcome (dependent) variable?  (2) Which variables in particular are significant predictors of the outcome variable, and in what way do they–indicated by the magnitude and sign of the beta estimates–impact the outcome variable?  These regression estimates are used to explain the relationship between one dependent variable and one or more independent variables.  The simplest form of the regression equation with one dependent and one independent variable is defined by the formula y = c + b*x, where y = estimated dependent variable score, c = constant, b = regression coefficient, and x = score on the independent variable.

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Create synthetic data for Linear Regression
np.random.seed(0)
X_linear = np.sort(5 * np.random.rand(80, 1), axis=0)
y_linear = 2 * X_linear + 1 + np.random.randn(80, 1)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_linear, y_linear)
y_linear_pred = linear_model.predict(X_linear)
```

```python
plt.figure(figsize=(12, 6))

# Linear Regression Plot
plt.subplot(1, 2, 1)
plt.scatter(X_linear, y_linear, color='blue', label='Data')
plt.plot(X_linear, y_linear_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
mse_linear = mean_squared_error(y_linear, y_linear_pred)
print(f"Mean Squared Error (Linear Regression): {mse_linear:.2f}")
```

## Non-Linear Regression

Nonlinear regression is a form of regression analysis in which data fit to a model is expressed as a mathematical function. Nonlinear regression fits a nonlinear model to the data. It is useful for describing phenomena such as growth rates, which often accelerate or decelerate and eventually reach a plateau. Nonlinear regression models are anything that doesn’t follow this one form. For example, a model that’s a polynomial of order greater than one is nonlinear regression. A model that’s an exponential or logarithmic function is also nonlinear regression.

```python
# Create synthetic data for Non-Linear Regression (Polynomial Regression)
X_nonlinear = np.sort(5 * np.random.rand(80, 1), axis=0)
y_nonlinear = 1 + 0.2 * X_nonlinear + 0.1 * X_nonlinear**2 + np.random.randn(80, 1)
```

```python
# Non-Linear (Polynomial) Regression
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X_nonlinear)
nonlinear_model = LinearRegression()
nonlinear_model.fit(X_poly, y_nonlinear)
y_nonlinear_pred = nonlinear_model.predict(X_poly)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.scatter(X_nonlinear, y_nonlinear, color='blue', label='Data')
plt.plot(X_nonlinear, y_nonlinear_pred, color='red', label='Non-Linear (Polynomial) Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Non-Linear (Polynomial) Regression')
plt.legend()

plt.show()
```

```python
mse_nonlinear = mean_squared_error(y_nonlinear, y_nonlinear_pred)
print(f"Mean Squared Error (Non-Linear Regression): {mse_nonlinear:.2f}")
```
