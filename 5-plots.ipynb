**Aim**: WAP to plot a horizontal and vertical bar plot using python. Change the color of each bar also.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Categories': ['A', 'B', 'C', 'D', 'E'],
    'Values': [20, 35, 10, 45, 30],
    'Colors': ['red', 'green', 'blue', 'orange', 'purple']
}

# Create a pandas DataFrame
df = pd.DataFrame(data)
```

# Vertical bar plot

```python
plt.figure(figsize=(8, 6))
plt.bar(data['Categories'], data['Values'], color=data['Colors'])
plt.title('Vertical Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# Horizontal bar plot

```python
plt.figure(figsize=(8, 6))
plt.barh(data['Categories'], data['Values'], color=data['Colors'])
plt.title('Horizontal Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
